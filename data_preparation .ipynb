```python
country_data=pd.read_excel("ISO 3166-1 alpha-2.xlsx")
```

```python
len(country_data["alpha-2"])
```

```python

for i in range(len(trade_data["Importer"])):
    for e in range(len(country_data["alpha-2"])):
        if trade_data["Importer"][i]==country_data["alpha-2"][e]:
            #im_country.append(country_data["name"].values[e])
            trade_data["Importer"]=trade_data["Importer"].replace(trade_data["Importer"][i],country_data["name"].values[e])
```

```python
trade_data
```

```python
for i in range(len(trade_data["Exporter"])):
    for e in range(len(country_data["alpha-2"])):
        if trade_data["Exporter"][i]==country_data["alpha-2"][e]:
            #im_country.append(country_data["name"].values[e])
            trade_data["Exporter"]=trade_data["Exporter"].replace(trade_data["Exporter"][i],country_data["name"].values[e])
```
